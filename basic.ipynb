```python
# sequance of Hailstone 
# if n is 1 then the sequance ends
# if n is even then the next n of the sequance = n/2
# if n is odd then the next n of the sequance =(3+ n) +1

n = int(input("Enter a number: "))
print("n is ", n)

while not (n==1):              # same as n!=1
    if (n%2 == 0):
        n/=2
    else:
        n =(3*n)+1
    print(n)
```

### Pythonic 

```python
data = ["apple","banana","mango","kiwi"]

# Simple looping of list
for val in data:
    print(val)

# Looping with index
for i, val in enumerate(data):
    print(i, val)

# Looping reverse list
for val in reversed(data):
    print(val)
```

```python
data = ["apple","banana","mango","kiwi"]
colors = ["green", "yellow","orange", "green"]

# Create a dictionary of pairs
dc = dict(zip(data, colors))


# Loop dictionary
for key, val in dc.items():
    print(key, val)

```
